```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# carregando imagem original
im = Image.open('./len_std.jpg')
```

```python
# passando a imagem para formato numpy
im_arr = np.asarray(im)
```

```python
def convert(a, to='gray_scale', limiar=128):
  # conversão para escala de cinza usando a media dos canais de cor
  gray = np.mean(a, axis=-1).astype(np.uint8)
  if to == 'gray_scale':
    return gray
  elif to == 'binary':
    # criando array com valores zeros
    zeros = np.zeros_like(gray)
    zeros[gray > limiar] = 255  # passando cor como 255 onde no array original a cor é maior que 128
    return zeros
```

```python
im  # imagem original
```

```python
# em escala de cinza
imgray = convert(im_arr, to='gray_scale')
Image.fromarray(imgray)
```

```python
# escala binária
im_bin = convert(im_arr, to='binary')
Image.fromarray(im_bin)
```
